```python
r = 0.01

#Решаем уравнения относительно волатильности
d1 = np.zeros(len(data))
d2 = np.zeros(len(data))
BIDsigma = np.zeros(len(data))
ASKsigma = np.zeros(len(data))
resB = np.zeros(len(data))

#Описываем это уравнение
def f(x, S, K, r, T, B, Param):
    d1 = (np.log(S / K) + (r + (x**2) / 2) * T) / (x  * np.sqrt(T))
    d2 = d1 - x * np.sqrt(T)
    if B == 'CALL':
        res = (S * scipy.stats.norm.cdf(d1) - K * np.exp(-r * T) * scipy.stats.norm.cdf(d2) - Param)
    else:
        res = (-S * scipy.stats.norm.cdf(-d1) + K * np.exp(-r * T) * scipy.stats.norm.cdf(-d2) - Param)
    return res

for i in range(0, len(data)):
    K = df.loc[i,'Страйк']
    T = df.loc[i,'Время до экспирации (в годах)']
    B = df.loc[i,'Тип']
    S = df.loc[i, 'Цена базисного актива']
    BID = df.loc[i,'Цена_BID']
    ASK = df.loc[i,'Цена_ASK']
    x0BID = [0.16, 0.16, 0.17, 0.28, 0.25, 0.2, 
             0.2, 0.18, 0.18, 0.19, 0.2, 0.2, 
             0.2, 0.31, 0.28, 0.26, 0.25, 0.23, 
             0.22, 0.2, 0.19, 0.19, 0.19, 0.3,
             0.27, 0.27, 0.26, 0.25, 0.24, 0.23,
             0.23, 0.14]
    x0ASK = [0.17, 0.17, 0.20, 0.3, 0.26, 0.23, 
             0.19, 0.19, 0.19, 0.20, 0.30, 0.26, 
             0.2, 0.21, 0.32, 0.28, 0.26, 0.25, 
             0.23, 0.25, 0.25, 0.25, 0.22, 0.35, 
             0.35, 0.31, 0.30, 0.28, 0.25, 0.28, 
             0.25, 0.33]
    # Получаем массив волатильности BID
    BIDsigma[i] = optimize.newton(f, x0BID[i], args = (S, K, r, T, B, BID,), tol=1e-5, maxiter = 100)
    # Получаем массив волатильности ASK
    ASKsigma[i] = optimize.newton(f, x0ASK[i], args = (S, K, r, T, B, ASK,), tol=1e-5, maxiter = 100)
    #print(f(BIDsigma[i],S, K, r, T, B, BID)) #<---проверка 
    #print(f(ASKsigma[i],S, K, r, T, B, ASK))
#print('BID sigma: ',BIDsigma)                          #<--вывод массивов волатильности
#print('ASK sigma: ',ASKsigma)
```

```python
# 2-ой этап

# коэффициент бета
b = 1

# вспомогательные функции для SABRsigma
def a(T, p):
    return p[0] + (p[1] + p[2] * T/ p[3]) * np.exp(-T / p[3])

def v(T, p):
    return p[4] + (p[5] + p[6] * T / p[7]) * np.exp(-T / p[7])

def p_(T, p):
    return p[8] + (p[9] + p[10] * T / p[11]) * np.exp(-T / p[11])

def f(T):
    return S * np.exp(r * T)

def z(T, K, p):
    return (v(T, p) / a(T, p)) * ((f(T) * K) ** ((1 - b)/2)) * np.log(f(T) / K)

def xi(T, K, p):
    return np.log(np.fabs((np.sqrt(np.fabs(1 - 2 * p_(T, p) * z(T, K, p) + z(T, K, p) ** 2)) + z(T, K, p) - p_(T, p)) / (1 - p_(T, p))))

# Формула SABR для поверхности волатильности
#def SABRsigma(p, T, K):                 
 #   return (a(T, p) * (1 + (((1 - b) ** 2) * (a(T, p) ** 2) / (24 * ((f(T) * K) ** (1 - b))) + 0.25 * p_(T, p) * b * v(T, p) * a(T, p) / ((f(T) * K) ** ((1 - b)/2)) + (2-3*(p_(T,p)**2)) * (v(T,p)**2) / 24) ) * T) /((f(T) * K) ** ((1 - b)/2) * (1 + (1 - b) ** 2 / 24 * np.log(np.fabs(f(T) / K)) ** 2 + (1 - b) ** 4 / 1920 * np.log(np.fabs(f(T) / K)) ** 4)) * z(T, K, p) / xi(T, K, p)
    
def SABRsigma1(p, T, K):
    return (a(T, p) * (1 + (((1 - b) ** 2) * (a(T, p) ** 2) / (24 * ((f(T) * K) ** (1 - b))) + 0.25 * p_(T, p) * b * v(T, p) * a(T, p) / ((f(T) * K) ** ((1 - b)/2)) + (2-3*(p_(T,p)**2)) * (v(T,p)**2) / 24) ) * T)
def SABRsigma2(p, T, K):
    return ((f(T) * K) ** ((1 - b)/2) * (1 + (1 - b) ** 2 / 24 * (np.log(np.fabs(f(T) / K)) ** 2) + ((1 - b) ** 4 / 1920) * np.log(np.fabs(f(T) / K)) ** 4))
def SABRsigma(p, T, K):
    return SABRsigma1(p,T,K) / SABRsigma2(p, T, K) * z(T, K, p) / xi(T, K, p)

# Формула cуммы разностей квадратов

def SSS(p, T, K, Param1,Param2):
    arr = SABRsigma(p,T,K)
    res = np.zeros(len(T))
    for i in range(0, len(T)):
        if Param1[i] > arr[i]:
            res[i] = np.fabs(Param1[i] - arr[i])
        elif arr[i] > Param2[i]:
            res[i] = np.fabs(arr[i] - Param2[i])
        else:
            res[i] = 0
    return res

T = np.zeros(len(data))
K = np.zeros(len(data))
#S = np.zeros(len(data))

for i in range(0, len(data)):
        T[i] = df.loc[i,'Время до экспирации (в годах)']
        K[i] = df.loc[i,'Страйк']
        #S[i] = df.loc[i, 'Цена базисного актива']


# Cчитываем массив чисел Соболева

SobolevBook = px.load_workbook('Sobolev.xlsx', data_only = True)  # подружаем файл с исходными данными

dataSobolev = list(SobolevBook['Лист1'].values)[0:]

lbMID =    [-10, -10, -10, -10, -10, -10, -10, -10, -10, -10, -10,  -10]
rbMID =    [ 10,  10,  10,  10,  10,  10,  10,  10,  10,  10,  10,   10]


#инициализируем  pMID. Теперь это матрица
pMID = np.zeros((len(dataSobolev), len(dataSobolev[0])))

#находим все 15 массивов pMID
for par in range(0, len(dataSobolev)):
    buf = optimize.least_squares(SSS, x0 = dataSobolev[par], bounds = (lbMID,rbMID), args=(T, K, BIDsigma,ASKsigma))
    print(par)
    pMID[par] = buf.x

#Ищем наименьшую сумму квадратов разностей (на самом деле в квадрат смысла возводить нет)
means = np.zeros(len(dataSobolev))
#считаем средние значения в каждом массиве, выбираем наименьшее и находим индекс
minimum = 10000
index = 0
for i in range(0, len(dataSobolev)):
    means[i] = np.mean(SSS(pMID[i], T, K, BIDsigma, ASKsigma))
    if (means[i] < minimum) and (means[i] > 0):       #  && (means[i] > 0) убрать после поправки значений Соболева
        minimum = means[i]
        index = i

#сохраняем результат с наименьшей суммой разностей квадратов
print('\n',means[index],'\n')
print(pMID[index])
SABRsigmaMID = SABRsigma(pMID[index], T, K)


#print('MIDsigma')
#print(MIDsigma)
#print('\n')
print(np.fabs(BIDsigma - SABRsigmaMID))
#print('Абсолютная погрешность: ', np.mean(np.fabs(BIDsigma - SABRsigmaMID)))
```
